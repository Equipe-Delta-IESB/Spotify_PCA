```python
## Bibliotecas
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import os 
import zipfile
```

# Carregando Dados

```python
os.chdir("../00_Dados/")
os.listdir()
```

```python
## Extraindo dados zipados
with zipfile.ZipFile(os.listdir()[0],"r") as zip_ref:
    zip_ref.extractall()
```

```python
## Criando dicionário com Dataframes
bases = os.listdir()
dict_raw_data = {}
for i in range(1,6):
    try:
        dict_raw_data['df_'+bases[i][:-4]] = pd.read_csv(bases[i])
        print(f'{bases[i]} carregado com sucesso...')
        os.remove(bases[i])  ## Removendo .csv carregado do diretório

    except:
        print(f'ERRO AO CARREGAR {bases[i]}')
## Deletando variáveis
del bases

dict_raw_data.keys()
```

# Análise Exploratória dos Dados

```python
ax = df_data.hist(bins=25, grid=False, figsize=(12,12), color='#1DB954', zorder=2, rwidth=0.9)
```

```python
# Valores ausentes

df_data.isna().sum()
```

```python
# Dropar duplicados

df_data.drop_duplicates(subset=['name','tempo'],keep='first', inplace=True)
```

```python
df_data.shape
```

```python
# Dropar colunas que nao serao interessantes para analise

df_data.drop(columns=['explicit', 'id', 'key', 'mode','release_date','name','artists'],inplace=True)
```

```python
# Outliers duration_ms

df_data.duration_ms.describe()
```

```python
df_data['duration_ms'].quantile(0.99)
```

```python
df_data = df_data[df_data.duration_ms <= 680000]
```

```python
df_data.shape
```

```python
ax = df_data.hist(bins=25, grid=False, figsize=(12,12), color='#1DB954', zorder=2, rwidth=0.9)
```

```python
# Normalizar loudness, duration_ms, tempo

#df_data.loudness = (df_data.loudness - df_data.loudness.min())/(df_data.loudness.max() - df_data.loudness.min())

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_data[['loudness','duration_ms','tempo']] = scaler.fit_transform(df_data[['loudness','duration_ms','tempo']])

df_data
```

```python
# Matriz Corr

plt.figure(figsize=(12,12))
corr = sns.heatmap(df_data.corr(),vmin=-1,vmax=1, annot=True,cmap='Greens')
```
